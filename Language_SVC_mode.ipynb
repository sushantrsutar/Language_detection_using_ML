```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df1=pd.read_csv('my_dataframe')
df2=pd.read_csv('bulgarian.csv')

df4=pd.read_csv('Telugu5.csv')
df=pd.read_csv('Danish.csv')
df5=pd.read_csv('Georgian.csv')
df6=pd.read_csv('German.csv')
df7=pd.read_csv('Greek.csv')
df8=pd.read_csv('Italian.csv')
df12=pd.read_csv('Kannada.csv')
df14=pd.read_csv('punjabi.csv')

df16=pd.read_csv('Tatar.csv')
```

```python
df_final=pd.concat([df,df1,df2,df4,df5,df7,df8,df14,df16],axis=0)
df_final
```

```python
df_final=df_final.drop(columns=['ბუნება'])
df_final
```

```python
#method to clean english 
import re
def remove_special_sym1(row):
    pattern=re.compile('[^\w\s]')
    pattern1=re.compile('[\d]')
    text=str(row['Text'])
    text=re.sub(pattern,'',text)
    text=re.sub(pattern1,'',text)
    return text
```

```python
df_final['Text']=df_final.apply(remove_special_sym1,axis=1)
df_final.tail(30)
```

```python

#method to clean english 
import re
def remove_special_sym(df_english):
    pattern=re.compile('[^\w\s]')
    pattern1=re.compile('[\d]')
    df_english['Text']=re.sub(pattern,'',df_english['Text'])
    df_english['Text']=re.sub(pattern1,'',df_english['Text'])
    return df_english['Text']
```

```python
X=np.array(df_final['Text'])

X
```

```python


from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv=TfidfVectorizer()
X=cv.fit_transform(X)
X
```

```python

Y=df_final['Language']
Y=np.array(df_final['Language'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=2)
```

```python
from sklearn.svm import SVC
svc_model=SVC()
svc_model.fit(x_train,y_train)
```

```python
svc_model.score(x_test,y_test)*100
```

```python
y_pred=svc_model.predict(x_test)
y_pred
```

```python
import matplotlib.pyplot as plt
fg,ax=plt.subplots(figsize=(18,18))
from sklearn.metrics import plot_confusion_matrix
#f = plt.figure(figsize=(10,3))
cmd = plot_confusion_matrix(svc_model,x_test, y_test,cmap=plt.cm.Blues, normalize=None,xticks_rotation=65,ax=ax)
#plt.figure(figsize=(10,8))                             
plt.title('Confusion Matrix')
plt.ylabel('Actual label')
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
#cm = metrics.confusion_matrix(y_test, y_pred)
#print(cm)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print()

print("Precision:",metrics.precision_score(y_test, y_pred,average=None))
print()
print("Sensitivity:",metrics.recall_score(y_test, y_pred,average=None))
print()
print("F1_score:",metrics.f1_score(y_test, y_pred,average=None))
```
